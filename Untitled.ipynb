```python
img_np = np.array(img)  # Get the image data as a NumPy array
```

```python
# Split the image data into its Red, Green, and Blue channels
red_channel = img_np[:,:,0]
green_channel = img_np[:,:,1]
blue_channel = img_np[:,:,2]
```

```python
# Convert the image to a numpy array
pink = np.array(img)
```

```python
# Define a range of pink color in RGB
lower_pink = np.array([200, 160, 160])
upper_pink = np.array([255, 200, 200])
```

```python
# Filter the image to extract only the pink pixels
pink = pink[(pink[...,0] >= lower_pink[0]) & (pink[...,0] <= upper_pink[0]) &
           (pink[...,1] >= lower_pink[1]) & (pink[...,1] <= upper_pink[1]) &
           (pink[...,2] >= lower_pink[2]) & (pink[...,2] <= upper_pink[2])]
```

```python
# Calculate the mean color of the pink pixels
mean_color = np.mean(pink, axis=(0, 1))
```

```python
print("Mean color of pink pixels:", mean_color)
```
